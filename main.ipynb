```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('avocado.csv')
print(df)
```

```python
plt.hist(df['AveragePrice'])
plt.title('Average Price')
plt.xlabel('Dollars')
plt.ylabel('Frequency')
plt.show()
```

```python
# from matplotlib.pyplot import figure

# figure(figsize=(10, 8), dpi=80)
plt.hist(df['4046'], color='red')
plt.title('PLU 4046')
plt.xlabel('Number of Avocados Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['4225'], color='pink')
plt.title('PLU 4225')
plt.xlabel('Number of Avocados Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['4770'], color='green')
plt.title('PLU 4770')
plt.xlabel('Number of Avocados Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['Total Volume'], color='yellow')
plt.title('Total Volume')
plt.xlabel('Number of Avocados Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
lst = ['Unnamed: 0', 'AveragePrice', 'Total Volume', '4046', '4225', '4770', 'year']
corrMatrix = df[lst].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```
